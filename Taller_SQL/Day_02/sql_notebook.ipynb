```
#Facturas de Clientes de Brasil, Nombre del cliente, Id de factura, fecha de la factura y el pais de la factura

query = """
SELECT c.FirstName, i.InvoiceId, i.InvoiceDate, i.BillingCountry
FROM customers c
JOIN invoices i ON c.CustomerId = i.CustomerId
WHERE c.Country = 'Brazil'
"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Facturas de Clientes de Brasil*

query = """
SELECT i.*
FROM customers c
JOIN invoices i ON c.CustomerId = i.CustomerId
WHERE c.Country = 'Brazil'
"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Muestra cada factura asociada a cada agente de ventas con su nombre completo*

query = """
SELECT i.*, e.LastName || ' ' || e.FirstName as 'Nombre_empleado'
FROM employees e
JOIN customers c ON e.EmployeeId = c.SupportRepId
JOIN invoices i ON c.CustomerId = i.CustomerId
"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Muestra el nombre del cliente, el pais, el nombre del agente y el total

query = """
SELECT c.FirstName || ' ' || c.LastName as 'Nombre_cliente', c.Country, e.FirstName || ' ' || e.LastName as 'Nombre_empleado', it.Quantity
FROM customers c
JOIN employees e ON c.SupportRepId = e.EmployeeId
JOIN invoices i ON c.CustomerId = i.CustomerId
JOIN invoice_items it ON i.InvoiceId = it.InvoiceId
"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Muestra cada articulo de la factura con el nombre de la cancion

query = """
SELECT it.InvoiceItemid, t.Name
FROM invoice_items it
JOIN tracks t ON it.TrackId = t.TrackId

"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Muestra todas las conciones con su nombre, formato, album y genero

query = """
SELECT t.Name, mt.Name, a.Title, g.Name
FROM tracks t
JOIN media_types mt ON t.MediaTypeId = mt.MediaTypeId
JOIN albums a ON t.AlbumId = a.AlbumId
JOIN genres g ON t.GenreId = g.GenreId

"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Muestra cuantas canciones hay en cada playlist y el nombre de cada playlist

query = """
SELECT COUNT(pt.TrackId), p.Name
FROM tracks t
JOIN playlist_track pt ON t.TrackId = pt.TrackId
JOIN playlists p ON pt.Playlistid = p.Playlistid
GROUP BY "pt.Playlistid"

"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```
