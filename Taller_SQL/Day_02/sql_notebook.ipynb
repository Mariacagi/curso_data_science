```
#Facturas de Clientes de Brasil, Nombre del cliente, Id de factura, fecha de la factura y el pais de la factura

query = """
SELECT c.FirstName, i.InvoiceId, i.InvoiceDate, i.BillingCountry
FROM customers c
JOIN invoices i ON c.CustomerId = i.CustomerId
WHERE c.Country = 'Brazil'
"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Facturas de Clientes de Brasil*

query = """
SELECT i.*
FROM customers c
JOIN invoices i ON c.CustomerId = i.CustomerId
WHERE c.Country = 'Brazil'
"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Muestra cada factura asociada a cada agente de ventas con su nombre completo*

query = """
SELECT i.*, e.LastName || ' ' || e.FirstName as 'Nombre_empleado'
FROM employees e
JOIN customers c ON e.EmployeeId = c.SupportRepId
JOIN invoices i ON c.CustomerId = i.CustomerId
"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Muestra el nombre del cliente, el pais, el nombre del agente y el total

query = """
SELECT c.FirstName || ' ' || c.LastName as 'Nombre_cliente', c.Country, e.FirstName || ' ' || e.LastName as 'Nombre_empleado', i. Total
FROM customers c
JOIN employees e ON c.SupportRepId = e.EmployeeId
JOIN invoices i ON c.CustomerId = i.CustomerId
"""
mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Muestra cada articulo de la factura con el nombre de la cancion

query = """
SELECT t.Name, it.InvoiceLineId, i.invoiceId
FROM invoice_items it
JOIN tracks t ON it.trackId = t.trackId
JOIN invoices i ON i.invoiceId = it.invoiceId
"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Muestra todas las conciones con su nombre, formato, album y genero

query = """
SELECT t.Name "Nombre Cancion", mt.Name 'Formato', a.Title, g.Name
FROM tracks t
JOIN media_types mt ON t.MediaTypeId = mt.MediaTypeId
JOIN albums a ON t.AlbumId = a.AlbumId
JOIN genres g ON t.GenreId = g.GenreId

"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Muestra cuantas canciones hay en cada playlist y el nombre de cada playlist

query = """
SELECT COUNT(*), p.Name
FROM playlists p
JOIN playlist_track pt ON p.PlaylistId = pt.PlaylistId
GROUP BY p.PlaylistId
"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Muestra cuánto ha vendido cada empleado

query = """
SELECT e.FirstName, SUM(i.Total) "Ventas"
FROM employees e
JOIN customers c ON e.EmployeeId = c.SupportRepId
JOIN invoices i ON i.CustomerId = c.CustomerId
GROUP BY e.EmployeeId
"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Quien ha sido el agente de ventas que más ha vendido en 2009?

query = """
SELECT e.FirstName, SUM(i.Total) "Ventas"
FROM employees e
JOIN customers c ON e.EmployeeId = c.SupportRepId
JOIN invoices i ON i.CustomerId = c.CustomerId
WHERE i.InvoiceDate LIKE "2009%"
GROUP BY e.EmployeeId
ORDER BY Ventas DESC
LIMIT 1
"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```

```
# Quien es son los 3 grupos que más han vendido?

query = """
SELECT ar.Name, SUM(it.UnitPrice) "Ventas"
FROM artists ar
JOIN albums a ON a.artistId = ar.artistId
JOIN tracks t ON t.albumId = t.albumId
JOIN invoice_items it ON it.trackId = t.trackId 
GROUP BY ar.ArtistId
ORDER BY Ventas DESC
LIMIT 3
"""

mi_def = pd.read_sql_query(query, conn)
mi_def
```
