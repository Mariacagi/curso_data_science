```python
%%time
pokemon_api_url = "https://pokeapi.co/api/v2/pokemon"
poke_json = requests.get(pokemon_api_url).json() #Esta es la primera peticion, la que lo transforma a json
pokemon = [] #Aqui lo va guardando

# Almacenamos 40 pokemons realizando peticiones hasta que la lista de pokemons tenga 40 o más pokemons
while len(pokemon) < 40:
    pokemon.extend(poke_json["results"]) #Lo que vamos a añadir a pokemon es lo que haya recibido en pokejson en la parte de results,es decir, el name y la url
    next_url = poke_json["next"] #Aquí se guarda la url siguiente
    poke_json = requests.get(next_url).json() #Aquí es donde hace la siguiente petición
print(len(pokemon))
pokemon
```

```python
diccionario_pokemons_url = {}
```

```python
diccionario_pokemons_url = {}
for i, e in enumerate(pokemon):
    nombre_pokemon = pokemon[i]["name"] #Uno a uno va ir accediando a los pokemon y añadiendo la clave para entrar en ella
    url_pokemon = pokemon[i]["url"] #Uno a uno va ir accediando a las url y añadiendo la clave para entrar en ella
    diccionario_pokemons_url[nombre_pokemon] = url_pokemon #Va a añadir el nombre del pokemon y como valor la url

diccionario_pokemons_url
```

```python
import pandas as pd 
df = pd. DataFrame(diccionario_pokemons_url, index=[0]) #Si lo ponemos sin index da error
df
```
