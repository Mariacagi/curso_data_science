```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

import warnings
warnings.filterwarnings('ignore')
```

```python
df1.info()
```

```python
df1["latitude"] = df1["latitude"].str.replace('%', '').astype(float)
df1
```

```python
df1['acq_date'] = df1['acq_date'].astype('datetime64[ns]')
df1['month'] = pd.DatetimeIndex(df1["acq_date"]).month

df2['acq_date'] = df2['acq_date'].astype('datetime64[ns]')
df2['month'] = pd.DatetimeIndex(df2["acq_date"]).month

df3['acq_date'] = df3['acq_date'].astype('datetime64[ns]')
df3['month'] = pd.DatetimeIndex(df3["acq_date"]).month

df4['acq_date'] = df4['acq_date'].astype('datetime64[ns]')
df4['month'] = pd.DatetimeIndex(df4["acq_date"]).month
```

## PRIMER PAR DE DATASET M6

```python
le = preprocessing.LabelEncoder()
```

```python
cat_col = ["satellite", "instrument", "daynight" ]
for i in cat_col:
    df1[i] = le.fit_transform(df1[i])
```

```python
# no cojo index, acq_date, acq_time y version
X = np.array(df1[["latitude", "longitude", "brightness", "scan", "track", "acq_time", "satellite", "instrument", "confidence", "bright_t31", "frp", "daynight", "month"]])
y = np.array(df1['fire_type'])
```

```python
#Divido en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = linear_model.LogisticRegression(n_jobs=-1)

model.fit(X_train, y_train)

score = model.score(X_test,y_test)*100
score
```

```python
model.fit(X, y)
```

```python
cat_col = ["satellite", "instrument", "daynight" ]
for i in cat_col:
    df3[i] = le.fit_transform(df3[i])
```

```python
predictions_submit = model.predict(df3[["latitude", "longitude", "brightness", "scan", "track", "acq_time", "satellite", "instrument", "confidence", "bright_t31", "frp", "daynight", "month"]])
predictions_submit
```

```python
df3["fire_type_pred"] = predictions_submit
```

```python
df3.to_csv("output/df3_maria.csv", index = False)
```
