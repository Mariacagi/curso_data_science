```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
lista_alumnos = [[29,  1.66], [22, 1.83], [31, 1.80], [27, 1.75], [22, 1.82], [35, 1.79], [34, 1.90], [26, 1.78], [26, 1.64], [23, 1.74],[24, 1.85], [28, 1.75], [23, 1.83],[32, 1.74], [34, 1.57], [27, 1.58], [38,1.70], [22, 1.79]]
```

```python
lista_edad = []
lista_altura = []
for x in lista_alumnos:
    lista_edad.append(x[0])
    lista_altura.append(x[1])

print(lista_edad)
print(lista_altura)
```

```python
X = np.array(lista_edad).reshape(-1, 1)
y = np.array(lista_altura).reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)

lm = LinearRegression(n_jobs=-1)
lm.fit(X_train, y_train)

print("\n\n####### Linear Regresion of edad vs altura #######\n")

print("####### Score train #######\n", lm.score(X_train, y_train) *100, "\n")
print("####### Score test #######\n", lm.score(X_test, y_test) *100, "\n")

plt.scatter(X, y, color='b')
plt.plot(X_test, lm.predict(X_test), color='red')

plt.show()
```

## Pregunta 1: ¿Es una buena técnica para este tipo de problemas? ¿Por qué? 

Sí, porque compara 2 columnas unicamente y se puede sacar una regresión lineal aunque para este caso faltan datos/ los datos no se pueden relacionar.
