### Se pide:

A partir del archivo "USA_Housing.csv", realizar la regresión lineal entre cada columna numérica con la columna 'price'. 

Cuando existe una columna(variable) que es la que se quiere comparar, se le llama 'target'. 

```python
# Importo los módulos para la parte de train y luego para crear la regresión lineal
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/USA_Housing.csv")
df
```

```python
# Defino las variables y el target
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Avg. Area Number of Bedrooms', 'Area Population']]
X1 = df['Avg. Area Income'].to_numpy()
Y = df["Price"].to_numpy()

X1 = X1.reshape(-1,1)
Y = Y.reshape(-1,1)
```

```python
# Divido entre train y test
X_train, X_test, Y_train, Y_test = train_test_split(X1, Y, test_size=0.2, random_state=101)
```

```python
# Creo el objeto de regresión lineal 
lm = LinearRegression(n_jobs=-1) # -1 para utilizar todos los procesadores
```

```python
# Entreno el Modelo
lm.fit(X_train, Y_train)
```

```python
# Veo el porcentaje de acierto en base a la parte de train
lm.score(X_train, Y_train) *100
```

```python
# Veo el porcentaje de acierto en base a la parte de test
lm.score(X_test, Y_test) *100
```

```python
# Muestro la regresión lineal entre la columna Avg. Area Income y la columna Price
plt.scatter(X1, Y, color='b')
plt.plot(X_test, lm.predict(X_test), color='red')

plt.show()
```

```python
lista_column = ['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Avg. Area Number of Bedrooms', 'Area Population']

def funcion_plot(size_of_test):

    for column in lista_column:
        X1 = df[column].to_numpy().reshape(-1,1)
        Y = df["Price"].to_numpy().reshape(-1,1)

        X_train, X_test, Y_train, Y_test = train_test_split(X1, Y, test_size=size_of_test, random_state=101)

        lm = LinearRegression(n_jobs=-1)
        lm.fit(X_train, Y_train)

        print("\n\n####### Linear Regresion of", column, "#######\n")

        print("####### Score train #######\n", lm.score(X_train, Y_train) *100, "\n")
        print("####### Score test #######\n", lm.score(X_test, Y_test) *100, "\n")

        plt.scatter(X1, Y, color='b')
        plt.plot(X_test, lm.predict(X_test), color='red')

        plt.show()

funcion_plot(size_of_test = 0.2)
```
